```python
import re
def normalize_title(title):
    # Remove years, special characters, and convert to lowercase
    title = re.sub(r'\(.*?\)', '', title)  # Remove anything in parentheses
    title = re.sub(r'[^\w\s]', '', title)  # Remove special characters
    return title.lower().strip()
```

```python
def add_movie_overviews_preserving_format(input_text, overview_dict):
    # Regular expression to capture movie titles with optional "(year)"
    movie_pattern = re.compile(r'"(.*?)\s*\(\d{4}\)?"')
    # Find all movie titles in the input text
    movies_found = movie_pattern.findall(input_text)
    
    # Function to get the overview for a given movie
    def get_overview(movie):
        movie = movie.strip()
        # Exact match first
        if movie in overview_dict:
            return f'{movie} the overview of {overview_dict[movie]}'
        else:
            # Try to find a partial match without the year
            for key in overview_dict.keys():
                if movie in key:
                    return f'{movie} the overview of {overview_dict[key]}'
        return f'{movie} the overview of not available'

    # Maintain original structure and replace only movie titles with overviews
    def replace_movies(match):
        movie = match.group(1).strip()
        overview_text = get_overview(movie)
        return f'"{overview_text}"'

    # Use regex sub to replace matches while preserving format
    modified_text = re.sub(r'"(.*?)\s*\(\d{4}\)?"', replace_movies, input_text)

    return modified_text
```

```python
for entry in data:
    entry['input']=add_movie_overviews_preserving_format(entry['input'],overview_dict)
```
