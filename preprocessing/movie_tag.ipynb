```python
data = pd.read_csv('../data/movie_10m/movies.csv')
data
```

```python
id_dict = {}

for i, row in data.iterrows():
    title = row['title']
    id_dict[title]= i+1
    
id_dict
```

```python
data2 = pd.read_csv('../data/movie_10m/tags.csv')
data2['tag']=data2['tag'].str.lower()
data2.head(2)
```

```python
for key, value in id_dict.items():
    lst = []
    if len(data2[data2['movieId'] == value])==0:
        id_dict[key]=['no']
    else:
        for entry in data2[data2['movieId']==value]['tag']:
            lst.append(entry)
        id_dict[key]=lst
```

```python
tag_dict = id_dict
```

```python
tag_dict
```

```python
import re
def normalize_title(title):
    # Remove years, special characters, and convert to lowercase
    title = re.sub(r'\(.*?\)', '', title)  # Remove anything in parentheses
    title = re.sub(r'[^\w\s]', '', title)  # Remove special characters
    return title.lower().strip()
```

```python
def match_title(row_title, tag_dict):
    normalized_row_title = normalize_title(row_title)
    row_title_words = set(normalized_row_title.split())
    
    for key in tag_dict.keys():
        normalized_key = normalize_title(key)
        key_words = set(normalized_key.split())
        
        #정확한 일치
        if normalized_row_title == normalized_key:
            return set(tag_dict[key])
        
        #부분일치 (title쪼개서 tag_dict에 모두 존재하는지)
        if row_title_words.issubset(key_words):
            return set(tag_dict[key])
        
        #별칭 매칭 확인
        if '(' in row_title:
            alias = re.search(r'\((.*?)\)', row_title)
            if alias:
                alias_text = alias.group(1).strip().lower()
                if alias_text in normalized_key:
                    return set(tag_dict[key])
    
    return ['no']  # Return empty list if no match found

# Apply function to genre_data DataFrame
genre_data['tag'] = genre_data['title'].apply(lambda x: ', '.join(map(str, match_title(x, tag_dict))))
```

```python
genre_tag_data = pd.read_csv('../data/movie_genre_tag/genre_tag_data.csv')
genre_tag_data
```

```python
import re
def normalize_title(title):
    # Remove years, special characters, and convert to lowercase
    title = re.sub(r'\(.*?\)', '', title)  # Remove anything in parentheses
    title = re.sub(r'[^\w\s]', '', title)  # Remove special characters
    title = title.lower().strip()
    title = re.sub(r'\s+', ' ', title)
    return title
```

```python
def add(input):
    prefer_pattern = r'User Preference:\s*"(.*?)"\n'
    prefer_match = re.search(prefer_pattern, input)
    prefer = []
    if prefer_match:
        prefer = prefer_match.group(1).split('", "')
        prefer = [movie.strip('"') for movie in prefer]
    
    unprefer_pattern = r'User Unpreference:\s*"(.*?)"\n'
    unprefer_match = re.search(unprefer_pattern, input)
    unprefer = []
    if unprefer_match:
        unprefer = unprefer_match.group(1).split('", "')
        unprefer = [movie.strip('"') for movie in unprefer]
    
    target_pattern = r'Whether the user will like the target movie\s*"(.*?)"'
    target = re.search(target_pattern, input)[1]
        
    like_movie = []
    for movie in prefer:
        #영화제목 정규화해서 검색
        genre_row = genre_tag_data.loc[genre_tag_data['title'].apply(normalize_title) == normalize_title(movie)]
        
        title = genre_row['title'].values[0]
        genre = genre_row['genre'].values[0]
        tag = genre_row['tag'].values[0]
        like_movie.append(f"{title} the genre of [{genre}] with [{tag}] tag")
        
    unlike_movie = []
    for movie in unprefer:
        #영화제목 정규화해서 검색
        genre_row = genre_tag_data.loc[genre_tag_data['title'].apply(normalize_title) == normalize_title(movie)]
        
        title = genre_row['title'].values[0]
        genre = genre_row['genre'].values[0]
        tag = genre_row['tag'].values[0]
        unlike_movie.append(f"{title} the genre of [{genre}] with [{tag}] tag")
    
    target_movie = ""
    if target:
        normalized_target = normalize_title(target)
        genre_row = genre_tag_data.loc[genre_tag_data['title'].apply(normalize_title) == normalized_target]

        title = genre_row['title'].values[0]
        genre = genre_row['genre'].values[0]
        tag = genre_row['tag'].values[0]
        target_movie = f"{title} the genre of [{genre}] with [{tag}] tag"
    
    # 업데이트된 입력 생성
    update_input = (
        f'User Preference: {" / ".join(like_movie)}\n' if like_movie else ''
    )
    update_input += (
        f'User Unpreference: {" / ".join(unlike_movie)}\n' if unlike_movie else ''
    )
    update_input += (
        f'Whether the user will like the target movie {target_movie}?' if target_movie else ''
    )
    
    return update_input
```

```python
for idx, entry in enumerate(data):
    entry['instruction'] = 'Given the user\'s preference and unpreference, along with the genres and tags of each movie, identify whether the user will like the target movie by answering "Yes." or "No."'
    entry['input'] = add(entry['input'])

    # Print progress after processing each entry
    print(f"Processed entry {idx+1}/{len(data)}")
```
